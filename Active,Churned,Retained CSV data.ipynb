```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from dateutil.relativedelta import relativedelta

#CSVs must have customer id as a number not scientific notation

#import two csv's to DataFrames
us_raw = pd.read_csv('test_orders.csv')
#can_raw = pd.read_csv('can_sales_june_21.csv').fillna(0)

raw = us_raw

#convert 'month' column to datetime
raw['Date']= pd.to_datetime(raw['month'])


#Group by correct columns
grouped_elements = ['total_sales','orders']
raw_grouped = raw.groupby(['month','customer_id','customer_name'])[grouped_elements].sum().astype(int).reset_index()




#find the number of new customers per month
df_first = raw_grouped.sort_values('month').groupby('customer_id').head(1)
new_customers = df_first.groupby('month').count()
sliced_new_customers = new_customers.iloc[:,0:1]



#Create crosstab() table 

raw_grouped_refresh = pd.crosstab(raw_grouped['customer_id'],raw_grouped['month']).astype(int).reset_index()
raw_grouped = pd.crosstab(raw_grouped['customer_id'],raw_grouped['month']).astype(int).reset_index()





actives = pd.DataFrame()
#for loop to add columns summing previous 12 months to get customers


start_col = 1
#companies active user period + 1
end_col = 4
num = end_col - start_col
loop_counter = 0



for x in range(len(raw_grouped.columns)-end_col+1) : 
    
    test_series = raw_grouped.iloc[:,start_col:end_col].sum(axis=1)
    new_col = test_series > 0
    date = raw_grouped.columns[end_col -1]
    start_col = start_col+1
    end_col = end_col+1
    num = num+1
    loop_counter = loop_counter +1
    

    actives[ str(date)] = new_col
        


        
#append a sum to the bottom of active customer df
active_cust_sum = actives[:].sum()
full = actives.append(active_cust_sum, ignore_index=True)
full_refresh = actives.append(active_cust_sum, ignore_index=True)
sliced = full.iloc[-1:,-len(raw_grouped.columns):]


monthly_cust_sum = raw_grouped[:].sum()

monthlys = raw_grouped.append(monthly_cust_sum,ignore_index=True)
monthly_customers = monthlys.iloc[-1:,-len(sliced.columns):]


#download to csv
sliced.to_csv('/Users/evandiazdearce/Documents/output/active_customersTotal.csv', index = False, header=True)
monthly_customers.to_csv('/Users/evandiazdearce/Documents/output/monthly_customers.csv', index = False, header=True)








####################################################


############## starting on Churned users ##################

cust_data = full
filt_cust_data = cust_data.iloc[:,len(raw_grouped_refresh.columns):]

churned_customers = pd.DataFrame()
active_int = actives.astype(int)




num_2 = 1
start_col_2 = 0
end_col_2 = start_col_2 +1 




for x in range(len(actives.columns)-1) : 
    
    new_col_2 = active_int.iloc[:,start_col_2] - active_int.iloc[:,end_col_2] > 0
    
    date_2 = actives.columns[num_2] 
    
    num_2 = num_2+1
    
    start_col_2 = start_col_2 +1
    end_col_2 = end_col_2 +1 
    
    churned_customers[str(date_2)] = new_col_2



    
cust_data_sum = -churned_customers[:].sum()


full_churn = churned_customers.append(cust_data_sum,ignore_index=True)
full_churn_sliced = full_churn.iloc[-1:,-len(sliced.columns)+1:]

display(full_churn)


full_churn_sliced.to_csv('/Users/evandiazdearce/Documents/output/churned.csv', index = False, header=True)



####################################################



########### starting retained Users #################


retained_customers = pd.DataFrame()


loop_counter_2 = 0
num_3 = 1



#def my_fun (var1,var2):
    #retained_customers[num_4] = np.where((active_int.iloc[:,var1]) and (raw_grouped[[var2]]) == 1, 1, 0)
    #return df



for x in range(len(active_int.columns)-1) : 
    
    
    
    search_month = active_int.columns[num_3]
    
    #check if customer was active for the previous month
    condition_1 = active_int.iloc[:,loop_counter_2]
    #check if customer placed an order for the month
    condition_2 = raw_grouped.loc[:,search_month]
    
    
    test_col = condition_1 + condition_2
    new_col_3 = test_col == 2


    
    #print(retained_cust[raw_grouped.columns[len(raw_grouped_cross.columns)-1+num_3]])    
    date_3 = search_month
    loop_counter_2 = loop_counter_2+1
    retained_customers[str(date_3)] = new_col_3
    num_3 = num_3+1 


retained_sum = retained_customers[:].sum() 
full_retained = retained_customers.append(retained_sum, ignore_index = True)
sliced_retained = full_retained.iloc[-1:,-len(actives.columns):]


sliced_retained.to_csv('/Users/evandiazdearce/Documents/output/retained_cust.csv', index = False, header=True)

print('ran')


###############################

display(monthly_customers)
display(sliced_retained)
rescued_customers = monthly_customers - sliced_retained

###################################


combined = sliced.append(full_churn_sliced.append(sliced_retained.append(monthly_customers)))
combined_t = combined.T

combined_t = combined_t.join(sliced_new_customers, how='left')



#Rename the columns to be easier to undertand
col_names = ['Active','Churned','Retained','Monthly','New']
combined_t.columns = col_names

#create columns for rescued, new+res/Churn, Monthly Offset (for the retention rate calc), and Retention Rate
combined_t['Rescued'] = combined_t['Monthly'] - combined_t['New'] - combined_t['Retained']
combined_t['New+Res/Churn'] = (combined_t['New']+combined_t['Rescued'])/-combined_t['Churned']
combined_t[['Monthly_Offset']] = combined_t[['Monthly']].shift(periods=1)
combined_t['Retention Rate'] = combined_t['Retained']/combined_t['Monthly_Offset']


#replace infinite and NaN's with 0's 
combined_t.replace([np.inf, -np.inf,'Nan'], np.nan, inplace=True)
combined_t['New'].fillna(0)
combined_t = combined_t.replace(np.nan, 0)

#create first x-axis
_, ax = plt.subplots()
#create second x-axis
ax2 = ax.twinx()


#create the 2 line plots for retention and new+res/Churn
ax2.plot(combined_t[['Retention Rate']], color='green',label= 'Retention Rate')
ax2.plot(combined_t[['New+Res/Churn']], color='purple',label = 'New+Res/Churn')
#set second Y axis limits
ax2.set_ylim([-3,3])


#create bar charts for churned, New, and Rescued
combined_t[['Churned']].plot(kind='bar', ax=ax,color='black')
combined_t[['New','Rescued']].plot(kind='bar',color=['lightskyblue','red'], ax=ax,stacked=True)
#remove legend so we can add it back as 1 legend with bars and lines
ax.get_legend().remove()
ax.grid()

#create new legend for all of the plots and anchor it outside of the plot 
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0,bbox_to_anchor=(1.5, 1))



display(combined_t)
plt.show()





```
